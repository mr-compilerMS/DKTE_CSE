# Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
# to analyze the relation between tip and bill using scatter plot
sns.relplot(x='tip',y='total_bill',data=tips)
```

```python
# line plot 
sns.relplot(x='tip',y='total_bill',data=tips,kind='line')
```

```python
# relation between time and tips
sns.relplot(x='total_bill',y='tip',data=tips,hue='time')
```

```python
# check if the client is smoker or not
sns.relplot(x='total_bill',y='tip',data=tips,style='smoker',hue='time',s=100)
```

```python
sns.catplot(x='tip',y='sex',data=tips , hue='time')
```

```python
sns.catplot(x='tip',y='sex',data=tips,jitter=False , hue='time')
```

```python
sns.catplot(x='tip',y='sex',data=tips,kind='box')
```

```python
sns.catplot(x='time',y='tip',data=tips,kind='bar')
```

```python
sns.catplot(x="day",y="total_bill",hue="smoker",kind="swarm",data=tips)
```

```python
sns.pairplot(data=tips,hue='time')
```
