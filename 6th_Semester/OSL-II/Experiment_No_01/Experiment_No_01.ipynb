```python
import pandas as pd
```

```python
df=pd.read_csv('Iris.csv')
df.head()
```

```python
df.set_index('Id',inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
zero_fill=df.fillna(0)
zero_fill.head()
```

```python
#fill with specific value for every column
df.fillna({'SepalLengthCm':1,'SepalWidthCm':2,'PetalLengthCm':3,'PetalWidthCm':4}).head()
```

```python
# fill using median values
df.fillna(df.median(numeric_only=True)).head()
```

```python
# forward fill
df.fillna(method='ffill').head()
```

```python
# backfill
df.fillna(method='bfill').head()
```

```python
df.fillna(df.mean(numeric_only=True)).head()
```

```python
df.fillna(method='bfill',inplace=True)
df.isna().sum()
```

```python
df.fillna(0,inplace=True)
```

```python
df.to_csv('IrisModified.csv')
```
