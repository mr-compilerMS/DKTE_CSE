## 1. Basics of Matplotlib

```python
from matplotlib import pyplot as plt
```

- Plotting Simple Values

```python
plt.plot([1,2,6,4] , [2,6,4,8])
plt.show()
```

- Plotting graph with axis names

```python
X=[10,32,54,60]
y=[43,55,23,56]
plt.plot(X,y)
plt.title("Random Values")
plt.ylabel("Y")
plt.xlabel("X")
plt.plot()
```

- Change line color

```python
from matplotlib import style
style.use('ggplot')
```

```python
plt.plot(X,y,'g', linewidth=5)
```

- ploting with grid

```python
plt.grid(True)
plt.plot(X,y)
```

```python
plt.bar([1,3,5,7,9],[10,100,32,43,60])
plt.show()
```

- Scatter plot

```python
price=[10,12,43,30,55,33]
sales=[100,89,120,43,33,55]
plt.scatter(price,sales,label="Sales",color="b")
plt.title("Scatter Plot")
plt.xlabel("Price")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

## 2. Seaborn

- Load some sample data

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('iris')
print(df.head())
print(df.describe())
```

- Histogram

```python
sb.distplot(df['petal_length'],kde=False)
plt.show()
```

- Kernel density estimation plot.

```python
sb.distplot(df['petal_length'],hist=False)
plt.show()
```

- Pairplot

```python
sb.pairplot(df[['sepal_length','sepal_width']],diag_kind = "kde",kind = "scatter")
```

- Regression Plot

```python
sb.regplot(x="petal_length",y="petal_width",data=df)
```
